```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sdv.timeseries import PAR
```

```python
from sdmetrics import load_demo
```

```python
orig_data = pd.read_csv("Multivariate_TS_Data.csv", parse_dates = ['DateTime'])
orig_data.head()
```

```python
orig_data.describe()
```

```python
sequence_index = 'DateTime'
model = PAR(sequence_index = sequence_index)
```

```python
model.fit(orig_data)
```

```python
new_data = model.sample(1)
new_data
```

```python
# Comparing original data to new data for each column

for i in new_data.columns[1:]:
    orig_stat = orig_data[[i]].describe().rename(columns = {'{}'.format(i): "Real Data - {}".format(i)})
    new_stat = new_data[[i]].describe().rename(columns = {'{}'.format(i): "Synthetic Data - {}".format(i)})
    display(orig_stat.join(new_stat))
```

```python
from sdmetrics.reports.single_table import QualityReport
from sdmetrics.timeseries import LSTMDetection
```

```python
metadata = {'fields': {'DateTime': {'type': 'datetime', 'format': '%Y-%m-%d'},
  'TimeSeries0': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries1': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries2': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries3': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries4': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries5': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries6': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries7': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries8': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries9': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries10': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries11': {'type': 'numerical', 'subtype': 'float'},
  'TimeSeries12': {'type': 'numerical', 'subtype': 'float'}},
 'constraints': [],
 'model_kwargs': {},
 'name': None,
 'primary_key': None,
 'sequence_index': 'DateTime',
 'entity_columns': [],
 'context_columns': []}

original_data = orig_data.copy()
original_data['DateTime'] = orig_data.index


# new_data['Entity'] = '-'
# original_data['Entity'] = '-'
# new_data = new_data.drop('Entity', axis = 1)

original_data.head()
```

```python
my_report = QualityReport()
my_report.generate(original_data, new_data, metadata = metadata)
```

```python
# calculate whether the synthetic data respects the min/max bounds
# set by the real data
from sdmetrics.single_column import BoundaryAdherence

BoundaryAdherence.compute(
    original_data['DateTime'],
    new_data['DateTime']
)
```

```python
# calculate whether the synthetic data is new or whether it's an exact copy of the real data
'''
Score
(best) 1.0: The rows in the synthetic data are all new. There are no matches with the real data.
(worst) 0.0: All the rows in the synthetic data are copies of rows in the real data.
'''

from sdmetrics.single_table import NewRowSynthesis

NewRowSynthesis.compute(
    original_data,
    new_data,
    metadata
)
```

```python
# LSTMDetection.compute(original_data, new_data, metadata = metadata)
# Does not work with fewer than 2 entities. 
# The least populated class in y has only 1 member, which is too few. The minimum number of groups for any class cannot be less than 2.
# So I'll need to unpivot into DateTime and Column for both and use the tagname column as entity to match demo example
```

```python
new_data_unpivot = pd.melt(new_data, id_vars = 'DateTime',value_vars = new_data.columns[1:])
original_data_unpivot = pd.melt(original_data, id_vars = 'DateTime',value_vars = new_data.columns[1:])

display(new_data_unpivot.head())
display(original_data_unpivot.head())
```

```python
flat_metadata = {'fields': {'DateTime': {'type': 'datetime', 'format': '%Y-%m-%d'},
  'variable': {'type': 'categorical'},
  'value': {'type': 'numerical', 'subtype': 'float'}},
 'constraints': [],
 'model_kwargs': {},
 'name': None,
 'primary_key': None,
 'sequence_index': 'DateTime',
 'entity_columns': ['variable'],
 'context_columns': []}

LSTMDetection.compute(original_data_unpivot, new_data_unpivot, metadata = flat_metadata)
```
